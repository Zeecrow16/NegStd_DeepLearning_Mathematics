```python
import numpy as np
import torch
```

```python
#generate random numbers
np.random.randn(5)
```

```python
#repeat after fixing the seed
np.random.seed(17)
print('Output: ')
print(np.random.randn(5))
print(np.random.randn(5))

# expected output:
# [ 0.27626589 -1.85462808  0.62390111  1.14531129  1.03719047]
# [ 1.88663893 -0.11169829 -0.36210134  0.14867505 -0.43778315]
```

##New seed mechanism in Numpy

```python
randseed1 = np.random.RandomState(17)
randseed2 = np.random.RandomState(20210530)

#Expected Output:
# [ 0.27626589 -1.85462808  0.62390111  1.14531129  1.03719047]
# [-0.24972681 -1.01951826  2.23461339  0.72764703  1.2921122 ]
# [ 1.88663893 -0.11169829 -0.36210134  0.14867505 -0.43778315]
# [ 2.171257    1.15231025 -1.81881234 -0.13804934  0.53983961]
#Not seeded value:
# [ 1.1191818  -0.98339611  0.24452002 -0.58140974  0.4295639 ]

print(randseed1.randn(5))
print(randseed2.randn(5))
print(randseed1.randn(5))
print(randseed1.randn(5))
print(np.random.randn(5))
```

##In PyTorch

```python
torch.randn(5)
```

```python
torch.manual_seed(17)

print('Output: ')
print('Pytorch: ', torch.randn(5))
print('Numpy: ',np.random.randn(5))
```
