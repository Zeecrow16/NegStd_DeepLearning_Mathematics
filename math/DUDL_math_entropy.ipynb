##Entropy

The original code here is wrong because it is only computing the probability for one event. Furthermore, there is no indication of the likelihood of the event not occuring.

```python
#probability of an event happening
p = 0.25

#NOT the correct formula:
H = - (p*np.log(p))
print(H)
```

Correct code:

```python
x = [.25, .75]
H = 0
for p in x:
  H -= p*np.log(p)
print(H)
```

##Cross-Entropy

```python
#All probs must sum to 1
p = [1, 0]
q = [.25, .75]

H = 0
for i in range(len(p)):
  H -= p[i]*np.log(q[i])
print(H)
```
