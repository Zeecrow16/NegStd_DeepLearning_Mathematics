```python
import numpy as np
import matplotlib.pyplot as plt
#sympy = symbolic math in Python
import sympy as sym
import sympy.plotting.plot as symplot
```

```python
#create symbolic variables in sympy
x = sym.symbols('x')

#create a function
fx = 2*x**2

#compute its derivate
df = sym.diff(fx, x)

#print them
print('function: ',fx)
print('derivative: ', df)
```

```python
#plot
symplot(fx, (x,-4,4), title='The function')
plt.show()
```

```python
#the derivative
symplot(df, (x,-4,4), title='The derivative')
plt.show()
```
