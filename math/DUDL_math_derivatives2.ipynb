```python
import numpy as np
import sympy as sym
from IPython.display import display
```

```python
#create symbolic variables in sympy
x = sym.symbols('x')

#create two functions
fx = 2*x**2
gx = 4*x**3 - 3*x**4

#compute their individual derivatives
df = sym.diff(fx)
dg = sym.diff(gx)

#apply the product rule
product_rule = df*gx + fx*dg
```

```python
#using sympy
sympy = sym.diff(fx*gx)
print('Functions: ')
display(fx)
display(gx)

print('Derivatives: ')
display(df)
display(dg)

print('Using sympy: ')
display(sympy)

print('Manual product rule: ')
display(product_rule)
```

```python
#chain rule

#create embedded function
fx = ( x**2 + 4*x**3)**5

print('Function: ')
display(fx)

df = sym.diff(fx)
print('Derivative: ')
display(df)
```
