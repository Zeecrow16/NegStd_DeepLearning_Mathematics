```python
import numpy as np
```

##Mean

```python
x = [1,2,4,6,5,4,0]
n = len(x)

#compute the mean
mean1 = np.mean(x)
mean2 = np.sum(x) / n

print(mean1)
print(mean2)
```

##Variance

```python
var1 = np.var(x)

#Below is copying the formula:
var2 = (1/(n-1)) * np.sum((x-mean1)**2)

#corrected version:
var3 = np.var(x, ddof=1)

print('Original version (incorrect): ',var1)
print('Formula: ',var2)
print('Corrected version: ',var3)
```
