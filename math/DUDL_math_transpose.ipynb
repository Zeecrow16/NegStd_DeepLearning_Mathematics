```python
import numpy as np
import torch
```

```python
#create a vector
nv = np.array([[1,2,3,4]])
print(nv), print(" ")

#transpose
print(nv.T), print(" ")

#Transpose again
nvT = nv.T
print(nvT.T)
```

```python
#Create a matrix
nM = np.array([[1,2,3,4], [5,6,7,8]])
print(nM), print(" ")

#transpose
print(nM.T), print(" ")

#Transpose again
nMT = nM.T
print(nMT.T)
```
