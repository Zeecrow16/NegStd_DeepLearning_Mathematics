```python
import numpy as np
```

```python
#create a vector
v = np.array([1, 40, 2, -3])
minval = np.min(v)
maxval = np.max(v)

print('Min, Max: %g,%g' %(minval, maxval))

#with argmin, argmax
minidx = np.argmin(v)
maxdix = np.argmax(v)

#remember the index starts at 0
print('Min, Max: %g,%g' %(minidx, maxdix))
```

```python
#with a matrix
M = np.array([[0,1,10],[20,8,5]])

print(M), print(' ')
```

```python
#minimum value
minval1 = np.min(M) #Minimun from entire matrix
minval2 = np.min(M, axis=0) #minimum of each column
minval3 = np.min(M, axis=1) #minimum for each row

print(minval1)
print(minval2)
print(minval3)
```

```python
#minimum index
minidx1 = np.argmin(M) #Minimun from entire matrix
minidx2 = np.argmin(M, axis=0) #minimum of each column
minidx3 = np.argmin(M, axis=1) #minimum for each row

print(minidx1)
print(minidx2)
print(minidx3)
```
