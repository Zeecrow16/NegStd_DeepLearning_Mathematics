```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
#Computing the softmax function in numpy
z = [1,2,3]

#Compute the the natural exponent
num = np.exp(z)
#Compute the denominator - the sum over all of the e to the z's.
den = np.sum(np.exp(z))
#Compute the sigma
sigma = num / den

#Output of the softmax function:
print(sigma)
print(np.sum(sigma))
```

```python
#Repeat with random integers
z = np.random.randint(-5, high=15, size=25)
print(z)

#Compute the softmax
num = np.exp(z)
den = np.sum(num)
sigma = num/den


#Compare
plt.plot(z, sigma, 'ko')
plt.xlabel('Original number (z)')
plt.ylabel('Softmax $\sigma$')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```

```python
#Here is the example using PyTorch

#Create an instance if the softmax activation class
soft = nn.Softmax(dim=0)

#Then apply the data to that function - convert z (list) into pytorch tensor.
sigmaT = soft(torch.Tensor(z))

#Results
print(sigmaT)
```

```python
plt.plot(z, sigmaT, 'ko')
plt.xlabel('Manual Softmax')
plt.ylabel('Pytorch nn.Softmax')
plt.title(f'The two methods correlate at r={np.corrcoef(sigma, sigmaT)[0,1]}')
plt.show()
```
